# **Importações**

```python
!pip install ultralytics

from ultralytics import YOLO
import matplotlib.pyplot as plt
from google.colab import files
from PIL import Image
```

# **Treinando o modelo YOLOv8**

```python
!git clone https://github.com/mariana-florencio/-Treinamento-da-Rede-YOLO-com-o-Dataset-Thermal-Dogs-and-People.git
%cd /content/-Treinamento-da-Rede-YOLO-com-o-Dataset-Thermal-Dogs-and-People

data_yaml_path = "/content/-Treinamento-da-Rede-YOLO-com-o-Dataset-Thermal-Dogs-and-People/data.yaml"

model = YOLO("yolov8n.pt")

model.train(
    data=data_yaml_path,
    epochs=50,
    imgsz=640,
    batch=16,
    device="cuda"
)
```

# **Testando o Modelo com uma Imagem**

```python
uploaded = files.upload()

image_path = list(uploaded.keys())[0]
img = Image.open(image_path)

plt.imshow(img)
plt.axis('off')
plt.show()

results = model(image_path)
result = results[0]
result.show()
```
